```python
import time
from colorama import Fore
def imprimir_texto_animado(text):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(0.02)
```

```python
from camel.agents import RolePlaying

mensaje_tarea = input("Insertar tarea a realizar:")

# mensaje_tarea = "Diseña un juego personalizado usando pygame"


print(Fore.YELLOW + f"Tarea original:\n{mensaje_tarea}\n")
sesion_role_playing = RolePlaying("Programador de Computadoras", "Jugador", mensaje_tarea)
print(Fore.CYAN + f"Tarea especificada:\n{sesion_role_playing.mensaje_tarea}\n")

limite_turnos_chat, n = 10, 0
mensaje_asistente, _ = sesion_role_playing.iniciar_chat()
while n < limite_turnos_chat:
    n += 1
    (mensaje_asistente, _, _), (mensaje_usuario, _, _) = sesion_role_playing.step(mensaje_asistente)
    imprimir_texto_animado(Fore.BLUE + f"Usuario IA:\n\n{mensaje_usuario.contenido}\n\n")
    imprimir_texto_animado(Fore.GREEN + f"Asistente IA:\n\n{mensaje_asistente.contenido}\n\n")
    if "<CAMEL_TAREA_HECHA>" in mensaje_usuario.contenido:
        break
```
